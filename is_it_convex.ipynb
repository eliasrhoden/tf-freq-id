```python
import tf_freq_id 
import numpy as np
import matplotlib.pyplot as plt

ws = [1,100,500,1000]


taus = np.linspace(1/100,1)



for w in ws:
    mags = []
    phases = []
    for tau in taus:
        pt1 = tf_freq_id.PT1(tau)
        mag = pt1.get_log_mag(w)
        ph = pt1.get_phase(w)

        mags.append(mag)
        phases.append(ph)

    plt.figure(1)
    plt.plot(taus,mags,label=f"$\\omega=${w}")
    plt.xlabel('$\\tau$')
    plt.ylabel('db mag')
    plt.figure(2)
    plt.plot(taus,phases,label=f"$\\omega=${w}")
    plt.xlabel('$\\tau$')
    plt.ylabel('phase')

plt.legend()

```

```python
ws = [1,100,500,1000]

wcs = np.linspace(1,700,300)

delta = 0.7

for w in ws:
    mags = []
    phases = []
    for wc in wcs:
        pt2 = tf_freq_id.PT2(wc,delta)
        mag = pt2.get_log_mag(w)
        ph = pt2.get_phase(w)

        mags.append(mag)
        phases.append(ph)

    plt.figure(1)
    plt.plot(wcs,mags,label=f"$\\omega=${w}")
    plt.xlabel('$\\omega_c$')
    plt.ylabel('db mag')
    plt.title(f'$\delta={delta}$')
    plt.figure(2)
    plt.plot(wcs,phases,label=f"$\\omega=${w}")
    plt.xlabel('$\\omega_c$')
    plt.ylabel('phase')
    plt.title(f'$\delta={delta}$')

plt.legend()

```

```python
ws = [1,100,500,1000]

wc = 500

deltas = np.linspace(0.05,0.9,100)

for w in ws:
    mags = []
    phases = []
    for delta in deltas:
        pt2 = tf_freq_id.PT2(wc,delta)
        mag = pt2.get_log_mag(w)
        ph = pt2.get_phase(w)

        mags.append(mag)
        phases.append(ph)

    plt.figure(1)
    plt.plot(deltas,mags,label=f"$\\omega=${w}")
    plt.xlabel('$\delta$')
    plt.ylabel('db mag')
    plt.title(f'$\\omega_c={wc}$')
    plt.figure(2)
    plt.plot(deltas,phases,label=f"$\\omega=${w}")
    plt.xlabel('$\delta$')
    plt.ylabel('phase')
    plt.title(f'$\\omega_c={wc}$')

plt.legend()
```

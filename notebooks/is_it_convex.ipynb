```python
import numpy as np 
import matplotlib.pyplot as plt 



wn = np.linspace(1,100)

wis = [1,10,50,100,150]
dn = 0.9

for wi in wis:

    t1Re = (wn**2 - wi**2)
    t1Im = (2*wn*wi*dn)

    t1 = t1Re**2 + t1Im**2


    y = np.log10(np.sqrt(t1))


    plt.plot(wn,y)


# smaller deltas = less convexity = harder to find solution
```
